```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
```

```python
data = sns.load_dataset('titanic')
```

```python
#Handling NULL VALUES
```

```python
titanic = data.drop(columns = 'deck')
```

```python
titanic.dtypes
```

```python
titanic.info()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic.isnull().sum()
```

```python
data
```

```python
titanic['embarked'].unique()
```

```python
titanic = pd.get_dummies(titanic,columns = ['embarked','embark_town'])

#embarked_C (1 if the value was C, 0 otherwise)
#embarked_Q (1 if the value was Q, 0 otherwise)
#embarked_S (1 if the value was S, 0 otherwise)
```

```python
titanic.isnull().sum()
```

```python
# STATISTICAL DATA VISUALIZATION USING SEABORN
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
plt.title('Joint Plot')

#The central part scatter plot helps to visualize how these two variables relate to each other. 
#For instance, you can look for patterns like whether older passengers tend to pay higher fares or if there is no clear relationship between age and fare.

#On the top of the scatter plot, there's a histogram of 'fare', which shows the distribution of fare amounts across all passengers. 
#This helps you understand the spread of fares, typical fare values, and outliers.

#On the right side of the scatter plot, there's a histogram of 'age', showing the distribution of ages among the passengers. 
#This histogram helps you understand the age composition of the passengers, such as whether the passenger list skews younger or older, or if it's evenly distributed.
```

```python
#Scatter plot
plt.scatter(titanic['age'],titanic['fare'],marker='*',alpha=0.8)
plt.title('Scatter plot')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
#BOX PLOT
```

```python
sns.boxplot(x='pclass',y='age',data=titanic)
plt.title('Box Plot')
```

```python
sns.displot(x='age',data=titanic,bins=70)
# sns.set(rc={'figure.figsize':(5,5)})
plt.title('Distribution Plot')

#bar chart that represents the frequency distribution of a numeric variable. In this case, the variable is 'age'.
```

```python
sns.swarmplot(x='pclass',y='age',data=titanic)
plt.title('Swarm Plot')

#This type of plot is particularly useful when you need a detailed look at how numeric data is distributed across different categories without losing information about individual data points.
```

```python
sns.countplot(x='sex',data=titanic)
plt.title('Count Plot')
```

```python
tc = titanic.select_dtypes(include=[np.number]).corr()
sns.heatmap(tc,cmap='coolwarm')
plt.title('Heat Map')
```

```python
#Histogram Plot to check how the price of the ticket for each passenger is distributed
sns.histplot(x='fare',data=titanic)
# sns.set(rc={'figure.figsize':(5,5)})
plt.title('Histpgram Plot')
```

```python
#FACTOR PLOT
# Use catplot (formerly factorplot) to create a categorical plot
sns.catplot(x='survived', kind='count', hue='pclass', data=titanic)
plt.title('Count Plot for PClass Grouped by Survival')
plt.show()
```

```python
sns.catplot(x='survived',data=titanic,kind='count',hue='sex').set(title='catplot for Gender')
```
